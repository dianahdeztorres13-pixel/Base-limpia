```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Dannap7337/Pato/main/Base_limpiaclase.csv')
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
df.columns
```

```python
sns.kdeplot(df["CreditScore"], fill=True,color="Blue")
plt.title("Grafica de densidad de balance")
plt.show()
```

```python
sns.kdeplot(df["Salario"], fill=True,color="green")
plt.title("Grafica de densidad de balance")
plt.show()
```

```python
sns.kdeplot(df["Balance"], fill=True,color="orange")
plt.title("Grafica de densidad de balance")
plt.show()
```

```python
sns.kdeplot(df["Puntos"], fill=True,color="purple")
plt.title("Grafica de densidad de balance")
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.kdeplot(df['Balance'], fill=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribución de Balance')

sns.kdeplot(df['Salario'], fill=True, ax=axes[0, 1], color="red")
axes[0, 1].set_title('Distribución de Salario')

sns.kdeplot(df['CreditScore'], fill=True, ax=axes[1, 0], color="brown")
axes[1, 0].set_title('Distribución de CreditScore')

sns.kdeplot(df['Edad'], fill=True, ax=axes[1, 1], color="orange")
axes[1, 1].set_title('Distribución de Edad')

plt.tight_layout()

plt.show()
```

```python
#Para datos categoricos 
gender_counts = df['Genero'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90,colors=['Yellow', 'Orange'])
plt.title('Distribución por Género')
plt.show()
```

```python
#GRAFICOS DE VIOLIN
```

```python
sns.violinplot(x='CreditScore', data=df)

plt.title('Distribución de CreditScore')
plt.show()
```

```python
colu = [ 'CreditScore', 'Pais', 'Genero', 'Edad',  'Estadia', 'Balance', '#Prod', 'TarjetaCredito', 'Activo', 'Salario',  'Salida', 'Quejas', 'Nivel_satisf', 'Tipo_tarjeta', 'Puntos']
```

```python
for i in colu:
    # Gráfico de violín para CreditScore por Tipo de Tarjeta
    sns.violinplot(y=i, data=df)

    # Mostrar gráfico
    plt.title(f'Distribución de {i}')
    plt.show()
    
```
